## Data Overview 📊
Below is the sign-up data for three departments across a few events:

| Department   | Event          | Attendees |
|--------------|---------------|-----------|
| Engineering  | Escape Room    | 8 |
| Engineering  | Cooking Class  | 5 |
| Engineering  | Hackathon      | 10 |
| Marketing    | Escape Room    | 7 |
| Marketing    | Cooking Class  | 12 |
| Marketing    | Hackathon      | 4 |
| Support      | Escape Room    | 4 |
| Support      | Cooking Class  | 3 |
| Support      | Hackathon      | 6 |

```python
# Run this cell to set up the DataFrame
import pandas as pd

signups = [
    ('Engineering', 'Escape Room', 8),
    ('Engineering', 'Cooking Class', 5),
    ('Engineering', 'Hackathon', 10),
    ('Marketing', 'Escape Room', 7),
    ('Marketing', 'Cooking Class', 12),
    ('Marketing', 'Hackathon', 4),
    ('Support', 'Escape Room', 4),
    ('Support', 'Cooking Class', 3),
    ('Support', 'Hackathon', 6),
]

df = pd.DataFrame(signups, columns=['department', 'event', 'attendees'])
df
```

## Challenge Tasks 📝
We'll accomplish the following using **pandas**:
- **Group** by event to sum attendees.
- **Group** by department to see which teams participate the most.
- **Pivot** the data to create an event-by-department table.
- **Plot** the total attendees per event.

*Each step below contains instructions and an example of the expected output for verification.*

### Step 1: Total Attendees per Event
1. Use `groupby` on `event` and sum `attendees` into `event_totals`.
2. Display `event_totals`.

```python
event_totals = df.groupby('event')['attendees'].sum()
event_totals
```

### Step 2: Participation by Department
1. Use `groupby` on `department` and sum `attendees`.
2. Display the result.

```python
dept_totals = df.groupby('department')['attendees'].sum()
dept_totals
```

### Step 3: Pivot Table
1. Create a pivot table with events as rows and departments as columns showing attendee counts.
2. Display the pivot table.

```python
pivot = df.pivot_table(index='event', columns='department', values='attendees', fill_value=0)
pivot
```

### Step 4: Visualize
1. Plot a bar chart of `event_totals` using `.plot.bar()`.
2. Feel free to customize the plot.

```python
event_totals.plot.bar(title='Total Sign-ups per Event')
```
