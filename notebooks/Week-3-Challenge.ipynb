### 📦 Scenario: Snapshot Retail Inventory Revamp

> **Snapshot Retail** is a fast-growing e-commerce marketplace selling everything from gaming laptops to garden supplies.  
> Until now, every department exported its own CSV file and the warehouse team had to stitch them together by hand before feeding them into the stock-replenishment system. Errors, duplicate SKUs, and late updates are costing the company **thousands of dollars** every month.

#### 👩‍💻 Your Mission
> You’ve just joined the **Data Enablement Squad** as a junior Python developer. Your first ticket is to build a **clean, single-source Product Catalogue** object that downstream services (pricing, replenishment, search-indexing) can consume without extra parsing.

#### ⚠️ The Immediate Problem
> Each night the legacy ERP drops a *flat* list of records that looks like this:

```python
records = [
    ("Electronics", "Laptops", "Ultrabook 13",  12),
    ("Electronics", "Laptops", "Gaming 17",     7),
    ("Electronics", "Phones",  "Foldable X",    5),
    ("Home",        "Kitchen", "Air Fryer XL",  9),
    ("Home",        "Kitchen", "Blender Pro",   4),
    ("Home",        "Garden",  "Hose 50 m",     3),
]
```

#### ⚠️ Expected Output

> Downstream services, however, expect a hierarchical structure grouped by department → category → product, with the value being the current stock count:

```python
catalog = {
    "Electronics": {
        "Laptops":  {"Ultrabook 13": 12,
                     "Gaming 17":    7},
        "Phones":   {"Foldable X":   5},
    },
    "Home": {
        "Kitchen": {"Air Fryer XL": 9,
                    "Blender Pro":  4},
        "Garden":  {"Hose 50 m":    3},
    }
}
```

## 🎯 Deliverables & Acceptance Criteria

| Deliverable                         | Purpose                                                                    | Success Criteria                                                         |
|-------------------------------------|----------------------------------------------------------------------------|--------------------------------------------------------------------------|
| `build_catalog(records) -> dict`    | Produce a function that converts the flat list to the nested dictionary.        | ✅ Produces the exact nested structure shown in *Expected Output*.        |
| **Readability**                     | Future interns should grasp the code in under 5 minutes.                   | ✅ Clear variable names, inline comments, or docstring.                   |
| **Extensibility**                   | New sub-categories are coming next quarter.                                | ✅ Adding an extra hierarchy level requires minimal changes.              |
| **Zero third-party deps**           | Must run inside the company’s slim Python 3.12.8 container.                  | ✅ Uses only the Python Standard Library.                                |
